```python
import os, glob, shutil
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split

# 2. Paths
DATASET = "F:/smart-crossing/training/data/raw"  
OUT_DIR = "F:/smart-crossing/training/data/split"

os.makedirs(OUT_DIR, exist_ok=True)

# 3. Example: list images
images = glob.glob(os.path.join(DATASET, "images", "*.jpg"))
labels = glob.glob(os.path.join(DATASET, "labels", "*.txt"))
print("Found", len(images), "images")

# 4. Split train/val/test
train_imgs, test_imgs = train_test_split(images, test_size=0.1, random_state=42)
train_imgs, val_imgs  = train_test_split(train_imgs, test_size=0.2, random_state=42)

print("Train:", len(train_imgs), "Val:", len(val_imgs), "Test:", len(test_imgs))

# 5. Copy into split folders
def copy_files(img_list, split):
    img_out = os.path.join(OUT_DIR, split, "images")
    lbl_out = os.path.join(OUT_DIR, split, "labels")
    os.makedirs(img_out, exist_ok=True)
    os.makedirs(lbl_out, exist_ok=True)
    for img in img_list:
        name = os.path.basename(img)
        lbl  = os.path.join(DATASET, "labels", name.replace(".jpg",".txt"))
        shutil.copy(img, img_out)
        shutil.copy(lbl, lbl_out)

copy_files(train_imgs, "train")
copy_files(val_imgs, "val")
copy_files(test_imgs, "test")
```
