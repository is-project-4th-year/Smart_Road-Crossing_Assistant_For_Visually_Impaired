```python
import tensorflow as tf

MODEL_DIR = "F:/smart-crossing/training/checkpoints/mobilenet_ssd"
EXPORT_DIR = "F:/smart-crossing/edge/models/"

# 1. Load saved model
model = tf.saved_model.load(MODEL_DIR)

# 2. Convert to TFLite
converter = tf.lite.TFLiteConverter.from_saved_model(MODEL_DIR)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # quantization
tflite_model = converter.convert()

# 3. Save
with open(EXPORT_DIR + "detector.tflite", "wb") as f:
    f.write(tflite_model)

# 4. Write label map (same order as training)
with open(EXPORT_DIR + "labelmap.txt", "w") as f:
    f.write("\n".join(["zebra_crossing","person","car","bus","truck","traffic_light"]))
```
